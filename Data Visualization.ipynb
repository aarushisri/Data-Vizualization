# Line Chart

```python
df_spx_ftse_00_10 = pd.read_csv("C:/Users/AARUSHI/Downloads/line_chart_data.csv")
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
df_spx_ftse_00_10["new_date"] 
```

```python
labels = ["S&P500","FTSE100"]
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_00_10["new date"],df_spx_ftse_00_10["GSPC500"] )
plt.plot(df_spx_ftse_00_10["new date"],df_spx_ftse_00_10["FTSE100"] )
plt.title("S&P vs FTSE rETURNS(2000 - 2010)", fontsize=14,weight='bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels =labels, fontsize ="large")
plt.show()
```

 Note: 10 years is a long period and hence we have way to many datapoints in the chart. We need to increase the granularity of the chart hence we'll choose a smaller timeframe.So lets chart only the last two quarters of the year 2008(The Great Recession).For doing this, we have to slice our data.

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01')&
                                      (df_spx_ftse_00_10.new_date <= '2008-12-31') ]
df_spx_ftse_H2_08
```

```python
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_H2_08["new date"],df_spx_ftse_H2_08["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_H2_08["new date"],df_spx_ftse_H2_08["FTSE100"], color = "crimson" )
plt.title("S&P vs FTSE rETURNS(2000 - 2010)", fontsize=14,weight='bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels =labels, fontsize ="large")
plt.show()
```

# Scatter Plot

```python
df_real_estate = pd.read_csv("scatter_data.csv")
df_real_estate
```

```python
plt.figure(figsize =(12,8))
scatter = plt.scatter(df_real_estate['Area (ft.)'],
           df_real_estate['Price'],
           alpha = 0.6,
           c = df_real_estate['Building Type'],  #C represents colour
           cmap = 'viridis')  #to change the default color palette
plt.legend(*scatter.legend_elements(),   #marking the pointer
           loc = "upper left",
           title ="Building Type")
plt.title("Relationship between Area and Price of California Real Estate", fontsize = 14 , weight = 'bold')
plt.xlabel("Area (sq. ft)", weight ='bold')
plt.ylabel("Price (000's $)")
plt.show()
```

# Regression Analysis

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
df_ad_expenditure
```

```python
plt.figure(figsize=(10,8))
sns.regplot(x = "Budget",
            y = "Sales",
           data = df_ad_expenditure,
           scatter_kws ={"color" : "k"},
           line_kws ={"color" : "red"})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's units)")
plt.title("Effecte of Expenditure on Sales", fontsize = 14, weight ='bold')
plt.show()
```

```python
sns.lmplot(x="Budget",
          y = "Sales",
          data = df_ad_expenditure,
          height = 10,
          scatter_kws ={"color" : "k"},
           line_kws ={"color" : "red"})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's units)")
plt.title("Effecte of Expenditure on Sales", fontsize = 14, weight ='bold')
plt.show()
```
